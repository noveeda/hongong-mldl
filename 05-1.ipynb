```python
feature_names = ['alcohol', 'sugar', 'pH']
label_name = 'class'
data = wine[feature_names]
target = wine[label_name]
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 1. 시각화 설정 (이상치 제거 및 범위 확장)
plt.figure(figsize=(10, 6))

# whis=(0, 100): 수염을 0%~100% 지점(최소~최대)까지 연장
# showfliers=False: 이상치(동그라미)를 화면에 표시하지 않음
wine.drop('class', axis=1).boxplot(whis=(0, 100), showfliers=False)

plt.title('표준화(StandardScaler) 적용 전 피처별 범위 비교')
plt.ylabel('Value')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(
    data, target, test_size=0.2, random_state=42
)

print(train_input.shape, test_input.shape)
print(data)
```

```python
print(train_input.shape, test_input.shape)
print(data)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False

data = pd.DataFrame(wine[['alcohol', 'sugar', 'pH']])
df_scaled = pd.DataFrame(ss.transform(data), columns=data.columns)
plt.figure(figsize=(10, 6))

df_scaled.boxplot(whis=(0, 100), showfliers=False)

plt.title('표준화(StandardScaler) 적용 후 피처별 범위 비교')
plt.ylabel('스케일링된 값 (Z-score)')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 원본 데이터
data = pd.DataFrame(wine[feature_names])
# 스케일링한 데이터
df_scaled = pd.DataFrame(ss.transform(data), columns=data.columns)

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

for col in ["alcohol", "sugar", "pH"]:
    sns.kdeplot(data=data, x=col, ax=axes[0], label=col, fill=True, alpha=0.3)
    sns.kdeplot(data=df_scaled, x=col, ax=axes[1], label=col, fill=True, alpha=0.3)

for ax, title in zip(axes, ["스케일링 전", "스케일링 후"]):
    ax.legend(loc="upper right")
    ax.set_xlabel("Value")   # x축 라벨
    ax.set_ylabel("Density") # y축 라벨
    ax.set_title(title)      # 그래프 제목
plt.show()
```

```python
plt.figure(figsize=(12, 6))

# 3. 여러 컬럼의 분포를 겹쳐서 그리기 (KDE Plot)
# data=df_scaled를 통째로 넣으면 모든 수치형 컬럼을 자동으로 겹쳐줍니다.
sns.kdeplot(data=df_scaled, fill=True, alpha=0.3)

# 4. 표준화 확인을 위한 가이드라인 추가
plt.axvline(0, color='black', linestyle='--', linewidth=1, label='평균 (0)')

# 5. 그래프 꾸미기
plt.title('표준화 후 피처별 분포 비교 (Z-score Distribution)')
plt.xlabel('Z-score (표준편차 단위)')
plt.ylabel('밀도 (Density)')
plt.legend(title='Features')
plt.grid(axis='y', alpha=0.3)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# p = 양성 클래스 비율 (0 ~ 1)
p = np.linspace(0, 1, 200)

# gini = 1 - (p² + (1-p)²)
gini = 1 - (p**2 + (1-p)**2)

fig, ax = plt.subplots(figsize=(9, 5))
ax.plot(p, gini, color="#5B8DB8", lw=2.5)
ax.fill_between(p, gini, alpha=0.2, color="#5B8DB8")

# 주요 포인트
points = {
    "완벽 분류\n(gini=0.0)": (0.0, 0.0),
    "완벽 분류\n(gini=0.0)": (1.0, 0.0),
    "최대 불순도\n(gini=0.5)": (0.5, 0.5),
}

for label, (x, y) in points.items():
    ax.scatter(x, y, color="#E07B54", s=80, zorder=5)
    ax.annotate(label, xy=(x, y),
                xytext=(x + 0.05 if x < 0.5 else x - 0.22, y + 0.03),
                fontsize=9, color="#E07B54")

ax.axhline(0.5, color="#aaa", linestyle="--", lw=1, alpha=0.6, label="최댓값 0.5")
ax.axvline(0.5, color="#aaa", linestyle="--", lw=1, alpha=0.6)

ax.set_xlabel("양성 클래스 비율 (p)", fontsize=11)
ax.set_ylabel("Gini 불순도", fontsize=11)
ax.set_title("Gini 불순도 그래프", fontsize=14, fontweight="bold")
ax.set_xlim(0, 1)
ax.set_ylim(0, 0.6)
ax.legend(fontsize=9)
ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

p1 = np.linspace(0, 1, 100)
p2 = np.linspace(0, 1, 100)
P1, P2 = np.meshgrid(p1, p2)
P3 = 1 - P1 - P2

# p1+p2 <= 1 인 경우만 유효
mask = P3 >= 0
gini = np.where(mask, 1 - (P1**2 + P2**2 + P3**2), np.nan)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(P1, P2, gini, cmap='Blues', alpha=0.8)

ax.set_xlabel("p1")
ax.set_ylabel("p2")
ax.set_zlabel("Gini")
ax.set_title("3개 클래스 Gini 불순도 (3D)")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

depths = range(1, 200)
scores = {"gini": [], "entropy": [], "log_loss": []}

for criterion in ["gini", "entropy", "log_loss"]:
    for depth in depths:
        dt_test = DecisionTreeClassifier(criterion=criterion,
                                    max_depth=depth)
        dt_test.fit(train_scaled, train_target)
        scores[criterion].append(dt_test.score(test_scaled, test_target))

plt.figure(figsize=(10, 5))
for criterion, s in scores.items():
    plt.plot(depths, s, label=criterion, lw=2)

plt.xlabel("max_depth")
plt.ylabel("test score")
plt.title("gini vs entropy 깊이별 점수 비교")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

p = np.linspace(0.001, 0.999, 300)

gini    = 1 - (p**2 + (1-p)**2)
entropy = -(p*np.log2(p) + (1-p)*np.log2(1-p))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# 왼쪽: 전체 범위
for ax, xlim, title in zip(axes,
                            [(0, 1), (0, 0.3)],
                            ["전체 범위", "소수 클래스 구간 확대 (p < 0.3)"]):
    ax.plot(p, gini,    label="gini",    lw=2.5, color="#5B8DB8")
    ax.plot(p, entropy, label="entropy", lw=2.5, color="#E07B54")
    ax.set_xlim(xlim)
    ax.set_title(title)
    ax.set_xlabel("양성 클래스 비율 (p)")
    ax.set_ylabel("불순도")
    ax.legend()
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0.001, 1, 300)
y = np.log2(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y, color="#5B8DB8", lw=2.5)
plt.axhline(0, color="#aaa", lw=1, linestyle="--")
plt.axvline(1, color="#aaa", lw=1, linestyle="--")

plt.annotate("log₂(1) = 0", xy=(1, 0), xytext=(0.7, 0.5),
             fontsize=10, color="#000000",
             arrowprops=dict(arrowstyle="->", color="#E07B54"))

plt.annotate("x → 0 일수록\n-∞ 로 발산", xy=(0.05, -4.3), xytext=(0.2, -3.5),
             fontsize=10, color="#000000",
             arrowprops=dict(arrowstyle="->", color="#6BAE75"))

plt.xlabel("x (클래스 비율 p)")
plt.ylabel("log₂(x)")
plt.title("log₂ 함수 그래프")
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
